Algoritmo KNN

Este notebook é para exercitar algumas funcionalidades do algoritmo KNN. Será usado o "dataset" IRIS, presente na biblio "SKLEARN". 

Primeiramente, realizar a importação do dataset e do pandas.    

```python
from sklearn import datasets
import pandas as pd
```

```python
# carregango o dataset iris num objeto
iris = datasets.load_iris()     
```

```python
# criando um dataframe com o conjunto de dados iris
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
# criando um atributo alvo:

df['target'] = iris.target
```

```python
df # podemos observar que o atributo "target" apresenta um numero correspondente a cada tipo/classificação.
```

```python
# quais são os tipos de target_names (nomes da classificação)
iris.target_names # comando para verificar os nomes. Observe que são classificadas como: setosa, versicolor e virginica.
```

```python
# especificando, nomeando o que corresponde cada número do atributo 'target'
df['target_name'] = iris.target_names[df['target']]
df  # observa-se que: 0 = setosa;  1 = versicolor  e  2 = virginica.  O df tem 50 observações de cd planta
```

```python
# determinar quais features farão parte do modelo:
iris_features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
```

```python
# selecionadas as features, determinar o alvo (y) - "lembrando que precisa ser um atributo do tipo numérico para usar no algo KNN"
y = df.target
```

```python
y
```

```python
# do conjunto de dados do 'df', selecionar as features usadas (X) 'x em MAIÚSCULA p variáveis preditoras. 

X = df[iris_features]

# verificando X:
X
```

Uso do Scikit-learn

Define = escolher o modelo (definir os parametros).

Fit = Treinando o modelo.

Predict = realizando as previsões.

Evaluete = Análise do modelo.




```python
#importando o algo KNN

from sklearn.neighbors import KNeighborsClassifier
```

```python
# Selecionar o modelo
modelo = KNeighborsClassifier(3) # o 3 indica os 3 "vizinho mais próximo ao alvo"
```

```python
# etapa de treino:
modelo.fit(X,y)
```

```python
X
```

```python
y
```

```python
# uma vez treinado, vamos verificar os resultados preditos pelo modelo em questão

print(modelo.predict(X))
```

```python
# verificando a acurácia do modelo

modelo.score(X,y)
```
